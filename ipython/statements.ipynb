```python
from pyspark.sql import DataFrame
from pyspark.sql.functions import from_utc_timestamp, to_utc_timestamp

def toTimezone(dataFrame: DataFrame, tz: str) -> DataFrame:
    for colName, dtype in dataFrame.dtypes:
        if dtype == 'timestamp':
            # dataFrame = dataFrame.withColumn(colName, from_utc_timestamp(colName, tz))
            dataFrame = dataFrame.withColumn(colName, to_utc_timestamp(colName, tz))
    
    return dataFrame
```

```python
from pyspark.sql.functions import udf
from typing import Optional
import phonenumbers

@udf(returnType='string')
def parsePhoneNumber(number: Optional[str], region: str='US') -> Optional[str]:
    try:
        phoneNumber = phonenumbers.parse(number, region)
        return phonenumbers.format_number(phoneNumber, phonenumbers.PhoneNumberFormat.NATIONAL)
    except:
        return None
```

```python
%%time
from pyspark.sql.functions import col, length, regexp_replace, when

column = when(col('PhoneNumber').rlike(r'\(\d+\) \d+-\d+'), col('PhoneNumber')).otherwise(None)

(
    vendors
    .where('Phone is not null')
    .withColumn('PhoneNumber', regexp_replace('Phone', '^([^0-9]+)', ''))
    .withColumn('PhoneNumber', regexp_replace('PhoneNumber', '\s+(?=[^0-9])(.*)', ''))
    .withColumn('PhoneNumber', parsePhoneNumber('PhoneNumber'))
    .withColumn('PhoneNumber', column)
    .selectExpr(
        'ObjectID as ID',
        'Phone',
        'PhoneNumber'
    )
    .write
    .mode('overwrite')
    .parquet('/tmp/phoneNumbers')
    # .where('PhoneNumber is not null')
    # .where('length(PhoneNumber) != 14')
    # .sample(fraction=1e-2)
    # .show(truncate=False)
    # .toPandas()
    # .to_csv('/tmp/phoneNumbers.csv', index=False)
)
```
